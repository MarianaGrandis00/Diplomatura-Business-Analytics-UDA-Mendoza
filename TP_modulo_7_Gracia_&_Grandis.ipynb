# **Gráficos**

## **Desarrollar gráficos del DataFrame**

```
import pandas_datareader as dtr
```

```
df
```

```
clientes = df.groupby("ocupación").sum()
clientes.head()
```

```
clientes.reset_index(inplace=True)
clientes.head()
```

```
clientes.plot("ocupación","pdays", 
              kind="bar", 
              figsize=(16,8), 
              color=["green","lightblue"],
              grid=True, 
              title="Análisis de clientes", 
              ylabel = "pdays",
              xlabel = "ocupación");
```

```
clientes = clientes[clientes.duration > 10000]
clientes
```

```
clientes = clientes[clientes.duration > 10000]
```

```
clientes.plot("","duration",
              kind="pie",
              figsize=(10,10),
              labels = clientes.index,
              legend = "",
              ylabel = "",
              autopct="%0.1f %%",
              title="Analisis de clientes")
```
